```python
import os
header_list = os.listdir('headers')
header_include_list = []
for header in header_list:
    header_include_list.append('#include <' + header + '>\n')
```

```python
def rewriting(filename):
    with open(filename, 'r', encoding = 'UTF-8') as f1:
        lines = f1.readlines()
    n_lines = len(lines)
    for i in range(n_lines):
        if lines[i] in header_include_list:
            l = len(lines[i])
            header = lines[i][10 : l - 2]
            if filename[0 : 7] == 'headers':
                lines[i] = '#include "' + header + '"\n'
            else:
                lines[i] = '#include "../headers/' + header + '"\n'
    with open(filename, 'w') as f2:
        f2.writelines(lines)
```

```python
folder_names = os.listdir()
for folder in folder_names:
    if not '.' in folder:
        filename_list = os.listdir(folder)
        for file in filename_list:
            if '.' in file:
                filename = folder + '/' + file
                print(filename)
                rewriting(filename)
```
